# Decision Tree algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read input file
df = pd.read_csv("salaries.csv")
df.head()
```

```python
X = df.drop("salary_more_then_100k", axis = 1)
X
```

```python
y = df.salary_more_then_100k
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
X['company_n'] = le_company.fit_transform(X['company'])
X['job_n'] = le_job.fit_transform(X['job'])
X['degree_n'] = le_degree.fit_transform(X['degree'])
X.head()
```

```python
X_final = X.drop(['company', 'job', 'degree'], axis = 1)
X_final
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_final, y)
```

```python
model.score(X_final, y)
```

```python
model.predict([[2,2,0]])
```

```python
model.predict([[0,0,1]])
```

# Exercise - Titanic passenger servival 

```python
# Read input file 
df_titanic = pd.read_csv("titanic.csv")
df_titanic
```

```python
# columns - passengerid, name, sibsp, Parch, Ticket cabin , Embarked are not going to be used in the model study 
# create a input data with columns Pclass, Sex, Age, Fare - Independant variables 
# create a target data with column Survived - dependent variable 
Input = df_titanic.drop(['PassengerId', 'Survived', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns')
Input
```

```python
# missing_data = Input.isnull()
# missing_data.head(5)
Input.isnull().sum()

# Column age only contains NaN values , 177 which is significant in the total rows, 
# hence , instead of deleting rows, replace NaN values with mean of age
```

```python
# simply drop whole row with NaN in "Age" column
Input = Input.fillna(int(Input[['Age']].mean()))
Input

```

```python
# check for any NaN values in the dataframe
Input.isnull().sum()
```

```python
Input.shape
```

```python
Target = df_titanic.Survived
Target
```

```python
Target.isnull().sum()
```

```python
le_sex = LabelEncoder()
```

```python
Input['Sex_n'] = le_sex.fit_transform(Input['Sex'])
```

```python
Input.head()
```

```python
X_final = Input.drop(['Sex'], axis = 'columns')
X_final
```

```python
# train test split model
from sklearn.model_selection import train_test_split
X_final_train, X_final_test, y_train, y_test = train_test_split(X_final, Target, test_size = 0.2)
```

```python
len(X_final_train)
```

```python
len(X_final_test)
```

```python
from sklearn import tree
# model_titanic = tree.DecisionTreeClassifier()
model_titanic = tree.DecisionTreeClassifier()
```

```python
model_titanic.fit(X_final_train, y_train)
```

```python
model_titanic.score(X_final_train, y_train)
```

```python
model_titanic.predict(X_final_test)
```

```python
model_titanic.predict_proba(X_final_test)
```
